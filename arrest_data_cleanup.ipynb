```python
proj = pyproj.Proj({"init" : "epsg:3311"})
```

```python
arr_loc = csv_10_11_narc['Location'].values
lat_arr = []
long_arr = []
x_arr = []
y_arr = []
x_max = 170000
y_max = -407000
grid_size = 200
for loc in arr_loc:
    loc = loc.replace(' ','')
    loc = loc.replace('(','')
    loc = loc.replace(')','')
    lat,long = loc.split(',')
    x,y = proj(float(long),float(lat))
    lat_arr.append(lat)
    long_arr.append(long)
    x_arr.append(x)
    y_arr.append(y)

x_noise = grid_size*(0.5 - np.random.random_sample((len(x_arr,))))
y_noise = grid_size*(0.5 - np.random.random_sample((len(y_arr,))))

len(x_arr)
```

```python
csv_10_11_narc.insert(len(csv_10_11_narc.columns),'Lat',lat_arr,True)
csv_10_11_narc.insert(len(csv_10_11_narc.columns),'Long',long_arr,True)
```

```python
csv_10_11_narc.insert(len(csv_10_11_narc.columns),'X',x_arr,True)
csv_10_11_narc.insert(len(csv_10_11_narc.columns),'Y',y_arr,True)
```

```python
csv_10_11_narc.insert(len(csv_10_11_narc.columns),'X_noisy',x_arr + x_noise,True)
csv_10_11_narc.insert(len(csv_10_11_narc.columns),'Y_noisy',y_arr + y_noise,True)
```

```python
csv_10_11_narc_clean = csv_10_11_narc.loc[csv_10_11_narc['Time'] != 2400.0]
```

```python
b = map(lambda x: math.isnan(x),csv_10_11_narc_clean['Time'].values)
csv_10_11_narc_clean = csv_10_11_narc_clean.loc[~np.array(list(b))]
```

```python
arr_times = csv_10_11_narc_clean['Time'].values
iterer = map(lambda x: str(math.trunc(x)).rjust(4,'0'),arr_times)
arr_times_padded = list(iterer)
```

```python
arr_dates = csv_10_11_narc['Arrest Date'].values
arr_datetime = list(map(lambda x, y: str(x)+ ' ' +y, arr_dates, arr_times_padded))
csv_10_11_narc_clean.insert(1,'Datetime',arr_datetime)
```
